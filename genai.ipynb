```python
orders = {
    "ORD101": "Your order has been delivered successfully.",
    "ORD102": "Your order has been shipped and will arrive in 2 days.",
    "ORD103": "Your order is out for delivery.",
    "ORD104": "Your order is being processed.",
    "ORD105": "Your order has been cancelled."
}
```

```python
def order_delivery_status(order_id):
    order_id = order_id.upper()

    if order_id in orders:
        return orders[order_id]
    else:
        return "Sorry, we couldn't find your order. Please check your Order ID."
```

```python
while True:
    user_input = input("Customer (Enter Order ID or type exit): ")

    if user_input.lower() == "exit":
        print("Support Bot: Thank you for contacting support!")
        break

    response = order_delivery_status(user_input)
    print("Support Bot:", response)
```
